```python
import pandas as pd
```

```python
df = pd.read_pickle('kicking')
```

```python
columns = ['GP', 'FGM_A', 'LNG', 'XPM_A', 'PTS']
percents = ['FG%', 'XP%']
```

```python
df
```

```python
dash = df['FG%'][40]
val = dash
```

```python
for column in columns:
    for i in range(len(df[column])):
        item = df[column][i]
        if item == val:
            df[column][i] = '0'
        else:
            df[column][i] = item
```

```python
for column in columns:
    df[column] = df[column].map(lambda x: int(x))
```

```python
for column in percents:
    for i in range(len(df[column])):
        item = df[column][i]
        if item == val:
            df[column][i] = '0.0'
        else:
            df[column][i] = item
```

```python
for column in percents:
    for i in range(len(df[column])):
        item = str(df[column][i])
        ones, tenths = item.split('.')
        ones = int(ones)
        tenths = int(tenths) * .1
        df[column][i] = ones + tenths
```
