```python
columns = ['GP', 'STK', 'ATK', 'CMB', 'INT', 'YDS', 'LNG', 'TD', 'FMB', 'FREC', 'FTD', 'PDEF', 'SFTY']
```

```python
dash = df['LNG'][508]
val = dash
```

```python
for column in columns:
    for i in range(len(df[column])):
        item = df[column][i]
        if item == val:
            df[column][i] = '0'
        else:
            df[column][i] = item
```

```python
for column in columns:
    df[column] = df[column].map(lambda x: int(x))
```

```python
df['SCK'][299]
```

```python
for i in range(len(df['SCK'])):
    item = df['SCK'][i]
    if item == val:
        df['SCK'][i] = '0.0'
    else:
        df['SCK'][i] = item
```

```python
df['SCK'].value_counts()
```

```python
for i in range(len(df['SCK'])):
    item = str(df['SCK'][i])
    ones, tenths = item.split('.')
    ones = int(ones)
    tenths = int(tenths) * .1
    df['SCK'][i] = ones + tenths
```
